```python
from pyspark.sql.functions import col
from datetime import datetime
```

```python
clean_customer = spark.table("default.clean_customer")
hub_customer = spark.table("default.hub_customer")
sat_customer = spark.table("default.sat_customer")
curated_customer_orders = spark.table("default.curated_customer_orders")
```

```python
print("========== ROW COUNT VALIDATION ==========")

print("Clean Customer :", clean_customer.count())
print("Hub Customer :", hub_customer.count())
print("Satellite Customer :", sat_customer.count())
print("Curated Customer Orders :", curated_customer_orders.count())
```

```python
print("========== NULL VALIDATION ==========")

null_customer = clean_customer.filter(
    col("Customer_ID").isNull()
).count()

print("Null Customer_ID :", null_customer)
```

```python
print("========== DUPLICATE VALIDATION ==========")

duplicate_customer = (
    clean_customer
    .groupBy("Customer_ID")
    .count()
    .filter(col("count") > 1)
)

duplicate_customer.show()
```

```python
print("========== HASH KEY VALIDATION ==========")

hub_customer.select(
    "Customer_ID",
    "Customer_HK"
).show(10, truncate=False)
```

```python
print("========== SCD TYPE 2 VALIDATION ==========")

sat_customer.select(
    "Customer_Name",
    "Is_Current",
    "Effective_From",
    "Effective_To"
).show(truncate=False)
```

```python
validation_result = {
    "Execution_Time": str(datetime.now()),
    "Customer_Count": clean_customer.count(),
    "Hub_Count": hub_customer.count(),
    "Satellite_Count": sat_customer.count(),
    "Curated_Count": curated_customer_orders.count(),
    "Null_PK": null_customer,
    "Status": "SUCCESS"
}

print(validation_result)
```
