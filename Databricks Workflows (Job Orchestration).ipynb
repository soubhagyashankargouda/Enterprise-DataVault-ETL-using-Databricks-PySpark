```python
from datetime import datetime

print("=" * 50)
print("Customer DataVault ETL Job Started")
print("Start Time:", datetime.now())
print("=" * 50)
```

```python
tables = [
    "default.clean_customer",
    "default.hub_customer",
    "default.hub_order",
    "default.hub_product",
    "default.link_customer_order",
    "default.link_order_product",
    "default.sat_customer",
    "default.sat_order",
    "default.sat_product",
    "default.curated_customer_orders"
]

for table in tables:
    count = spark.table(table).count()
    print(f"{table} : {count} records")
```

```python
summary = []

for table in tables:
    df = spark.table(table)
    summary.append((table, df.count(), len(df.columns)))

summary_df = spark.createDataFrame(
    summary,
    ["Table_Name", "Row_Count", "Column_Count"]
)

summary_df.show(truncate=False)
```

```python
job_status = {
    "Job_Name": "Customer_DataVault_ETL",
    "Status": "SUCCESS",
    "Tables_Processed": len(tables),
    "Execution_Time": str(datetime.now())
}

print(job_status)
```

```python
print("=" * 50)
print("Customer DataVault ETL Job Completed Successfully")
print("End Time:", datetime.now())
print("=" * 50)
```
