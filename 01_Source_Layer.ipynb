```python
customer_df = spark.table("default.customer_1")
orders_df = spark.table("default.orders_1")
products_df = spark.table("default.products_1")
order_details_df = spark.table("default.order_details_1")
```

```python
customer_df.show(5)
```

```python
orders_df.show(5)
```

```python
products_df.show(5)
```

```python
order_details_df.show(5)
```

```python
customer_df.printSchema()

orders_df.printSchema()

products_df.printSchema()

order_details_df.printSchema()
```

```python
print("Customer Count :", customer_df.count())

print("Orders Count :", orders_df.count())

print("Products Count :", products_df.count())

print("Order Details Count :", order_details_df.count())
```

```python
from pyspark.sql.functions import col

customer_df.filter(col("Customer_ID").isNull()).show()

orders_df.filter(col("OrderID").isNull()).show()

products_df.filter(col("ProductID").isNull()).show()

order_details_df.filter(col("OrderID").isNull()).show()
```
