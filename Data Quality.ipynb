```python
from pyspark.sql.functions import *

raw_customer = spark.table("default.raw_customer")
raw_orders = spark.table("default.raw_orders")
raw_products = spark.table("default.raw_products")
raw_order_details = spark.table("default.raw_order_details")
```

```python
print("Customer Count :", raw_customer.count())
```

```python
raw_customer.filter(
    col("Customer_ID").isNull()
).show()
```

```python
raw_customer.groupBy("Customer_ID") \
    .count() \
    .filter(col("count") > 1) \
    .show()
```

```python
raw_customer.filter(
    ~col("Email").rlike("^[A-Za-z0-9+_.-]+@[A-Za-z0-9.-]+$")
).show()
```

```python
raw_customer.filter(
    (col("Name") == "") |
    col("Name").isNull()
).show()
```

```python
raw_orders.filter(
    col("OrderID").isNull()
).show()
```

```python
raw_orders.groupBy("OrderID") \
    .count() \
    .filter(col("count") > 1) \
    .show()
    
```

```python
raw_products.groupBy("ProductID") \
    .count() \
    .filter(col("count") > 1) \
    .show()
```

```python
raw_order_details.filter(
    col("ProductID").isNull()
).show()
```

```python
clean_customer = raw_customer.filter(
    col("Customer_ID").isNotNull() &
    col("Name").isNotNull()
)
```

```python
clean_customer.show()
```
