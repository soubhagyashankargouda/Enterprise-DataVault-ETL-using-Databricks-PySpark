```python
from pyspark.sql.functions import *

customer_df = spark.table("default.customer_1")
orders_df = spark.table("default.orders_1")
products_df = spark.table("default.products_1")
order_details_df = spark.table("default.order_details_1")

```

```python
raw_customer = customer_df \
    .withColumn("Load_Date", current_timestamp()) \
    .withColumn("Record_Source", lit("Customer_CSV")) \
    .withColumn("Batch_ID", lit(1))
```

```python
raw_customer.write.mode("overwrite").saveAsTable("default.raw_customer")
```

```python
raw_orders = orders_df \
    .withColumn("Load_Date", current_timestamp()) \
    .withColumn("Record_Source", lit("Orders_CSV")) \
    .withColumn("Batch_ID", lit(1))
```

```python
raw_orders.write.mode("overwrite").saveAsTable("default.raw_orders")
```

```python
raw_products = products_df \
    .withColumn("Load_Date", current_timestamp()) \
    .withColumn("Record_Source", lit("Products_CSV")) \
    .withColumn("Batch_ID", lit(1))
```

```python
raw_products.write.mode("overwrite").saveAsTable("default.raw_products")
```

```python
raw_order_details = order_details_df \
    .withColumn("Load_Date", current_timestamp()) \
    .withColumn("Record_Source", lit("OrderDetails_CSV")) \
    .withColumn("Batch_ID", lit(1))
```

```python
raw_order_details.write.mode("overwrite").saveAsTable("default.raw_order_details")
```
