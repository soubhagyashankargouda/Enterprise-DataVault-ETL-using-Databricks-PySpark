```python
import json

customer_config = {
    "source_table": "default.clean_customer",
    "target_table": "default.hub_customer",
    "primary_key": "Customer_ID",
    "record_source": "Customer_CSV",
    "load_type": "incremental"
}

print(json.dumps(customer_config, indent=4))
```

```python
source_table = customer_config["source_table"]
target_table = customer_config["target_table"]
primary_key = customer_config["primary_key"]
record_source = customer_config["record_source"]
load_type = customer_config["load_type"]

print(source_table)
print(target_table)
print(primary_key)
print(record_source)
print(load_type)
```

```python
df = spark.table(source_table)

df.show(5)
```

```python
df = spark.table(source_table)

print(f"Source Table : {source_table}")
print(f"Target Table : {target_table}")
print(f"Primary Key  : {primary_key}")
print(f"Load Type    : {load_type}")
print(f"Record Source: {record_source}")

df.show(5)
```

```python
row_count = df.count()

print(f"Total Records : {row_count}")
```

```python
from pyspark.sql.functions import col

null_pk = df.filter(col(primary_key).isNull())

print("Null Primary Key Records:", null_pk.count())
```

```python
duplicate_pk = (
    df.groupBy(primary_key)
      .count()
      .filter(col("count") > 1)
)

duplicate_pk.show()
```

```python
validation_result = {
    "Source_Table": source_table,
    "Target_Table": target_table,
    "Row_Count": row_count,
    "Null_PK": null_pk.count(),
    "Load_Type": load_type
}

print(validation_result)
```
